```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
```

```python
text2text_tkn = T5Tokenizer.from_pretrained("t5-small")
mdl = T5ForConditionalGeneration.from_pretrained("t5-small")
```

```python
def text2text_sentiment(text):
  inp = "sst2 sentence: "+text
  enc = text2text_tkn(inp, return_tensors="pt")
  tokens = mdl.generate(**enc)
  response = text2text_tkn.batch_decode(tokens)
  return response

para = "its fun to play sports"
out = text2text_sentiment(para)
```

```python
para = "the bear market is spoiling my party"
out = text2text_sentiment(para)
print(out)
```

# **Grammatical Acceptance**

```python
def text2text_acceptable_sentence(text):
  inp = "cola sentence: "+text
  enc = text2text_tkn(inp, return_tensors="pt")
  tokens = mdl.generate(**enc)
  response = text2text_tkn.batch_decode(tokens)
  return response

para = "the bear market is spoiling my party"
out = text2text_acceptable_sentence(para)
print(out)
```

# **Sentence Paraphrasing**

```python
def text2text_paraphrase(sentence1, sentence2):
  inp1 = "mrpc sentence1: "+sentence1
  inp2 = "sentence2: "+sentence2

  combined_inp = inp1 +" "+inp2

  enc = text2text_tkn(combined_inp, return_tensors="pt")
  tokens = mdl.generate(**enc)
  response = text2text_tkn.batch_decode(tokens)
  return response
```

```python
Sentence1 = "Kubernetes is an open-source container orchestration system for automating software deployment, scaling, and management. Originally designed by Google, the project is now maintained by a worldwide community of contributors, and the trademark is held by the Cloud Native Computing Foundation."
Sentence2 = "Kubernetes, also known as K8s, is an open source system for automating deployment, scaling, and management of containerized applications.It groups containers that make up an application into logical units for easy management and discovery. Kubernetes builds upon 15 years of experience of running production workloads at Google, combined with best-of-breed ideas and practices from the community."

Sentence1 = "Kubernetes is an open-source platform designed to automate deploying, scaling, and operating containerized applications."
Sentence2 = "Kubernetes automates the management of containerized applications, providing capabilities such as deployment, scaling, and operations orchestration for containers across clusters."

out = text2text_paraphrase(Sentence1,Sentence2)
print(out)
```

# **Is Sentence Deductible**

```python
def text2text_deductible(sentence1, sentence2):
  inp1 = "rte sentence1: "+ sentence1
  inp2 = "sentence2: "+ sentence2
  combined_inp = inp1 +" "+inp2
  enc = text2text_tkn(combined_inp, return_tensors="pt")
  tokens = mdl.generate(**enc)
  response = text2text_tkn.batch_decode(tokens)
  return response

sentence1 = "Securing Kubernetes may seem like a mystifying task"
sentence2 = "kubernetes security is mystifying"
out = text2text_deductible(sentence1, sentence2)
print(out)
```
